```python
def loadcsv(file):
    with open(file, 'r') as csv_file:
        ha = list(map(float, csv_file.readline().strip().split(',')))
    return ha


def load_pickle(file):
    """Load the datum pickle and returns as a dataframe

    Args:
        file (string): labels pickle from 247-decoding/tfs_pickling.py

    Returns:
        DataFrame: pickle contents returned as dataframe
    """
    with open(file, 'rb') as fh:
        datum = pickle.load(fh)

    return datum
```

### Plot Average Encoding for (no-shuffle data)

```python
def extract_correlations(directory_list):
    all_corrs = []

    for dir in directory_list:
        file_list = glob.glob(os.path.join(dir, '*'))

        for file in file_list:
            if any([item in file for item in ['EKG', 'ECG', 'SG']]):
                continue

            file_ext = os.path.splitext(file)[-1]
            if file_ext == '.csv':
                with open(file, 'r') as csv_file:
                    ha = list(map(float, csv_file.readline().strip().split(',')))
            elif file_ext == '.mat':
                ha = loadmat(file)['rc'].tolist()[0]
            else:
                print('invalid file format')
                
            all_corrs.append(ha)

    hat = np.stack(all_corrs)
    print(hat.shape)
    mean_corr = np.mean(hat, axis=0)
    return mean_corr, hat.shape[0]
```

```python
res_dir1 = '/scratch/gpfs/hgazula/podcast-encoding/results/20210401-no-shuffle/*'
res_dir2 = '/scratch/gpfs/hgazula/podcast-encoding/results/bobbi-matlab-no-perm/*'

mean_corr1, shape1 = extract_correlations(glob.glob(res_dir1))
mean_corr2, shape2 = extract_correlations(glob.glob(res_dir2))

print(mean_corr1.shape)
print(mean_corr2.shape)

if mean_corr1.shape[0] == 801:
    mean_corr1 = mean_corr1[320:-320]
    
if mean_corr2.shape[0] == 801:
    mean_corr2 = mean_corr2[320:-320]

fig, ax = plt.subplots()
lags = np.arange(-2000, 2001, 25)
ax.plot(lags, mean_corr1, 'k', label=f'harsha {shape1}')
ax.plot(lags, mean_corr2, 'r', label=f'bobbi {shape2}')
ax.legend()
ax.set(xlabel='lag (s)', ylabel='correlation', title='Average Encoding')
ax.grid()

fig.savefig("average_encoding.png")
plt.show()
```

```python
def return_electrode_files(subject_id, shuffle_flag='noshuffle'):
    if shuffle_flag == 'noshuffle':
        bobbi_661 = glob.glob(os.path.join(parent_dir, 'bobbi-matlab-no-perm', 'NY' + str(subject_id) + '*', '*'))
        harsha_661 = glob.glob(os.path.join(parent_dir, '20210401-no-shuffle', str(subject_id), '*'))
    elif shuffle_flag == 'shuffle':
        bobbi_661 = glob.glob(os.path.join(parent_dir, 'bobbi-matlab-perm', 'NY' + str(subject_id) + '*', '*'))
        harsha_661 = glob.glob(os.path.join(parent_dir, '20210401-phase-shuffle', str(subject_id), '*'))
    else:
        print('invalid shuffle flag')

    bobbi_661.sort()
    harsha_661.sort()
    
    bobbi_661 = [elec for elec in bobbi_661 if not any([item in elec for item in ['EKG', 'ECG', 'SG']])]
    harsha_661 = [elec for elec in harsha_661 if not any([item in elec for item in ['EKG', 'ECG', 'SG']])]
    
    if shuffle_flag == 'noshuffle':
        bobbi_661 = sorted(bobbi_661, key=lambda x: os.path.splitext(os.path.basename(x))[0].split('conversation1')[-1].split('.')[0])
        harsha_661 = sorted(harsha_661, key=lambda x: os.path.splitext(os.path.basename(x))[0].split('_comp')[0])
    elif shuffle_flag == 'shuffle':
        bobbi_661 = sorted(bobbi_661, key=lambda x: os.path.splitext(os.path.basename(x))[0].split('_perm')[0])
        harsha_661 = sorted(harsha_661, key=lambda x: os.path.splitext(os.path.basename(x))[0].split('_comp')[0])
    else:
        print('invalid shuffle flag')
        
    assert len(bobbi_661) == len(harsha_661)
    
    return (bobbi_661, harsha_661)


def print_electrodes(bobbi_folder, harsha_folder):
    
    bobbi_list = []
    for bobbi in bobbi_folder:
        bobbi_elec = os.path.basename(bobbi)
        bobbi_elec_name = bobbi_elec.split('conversation1')[-1].split('.')[0]
        bobbi_list.append(bobbi_elec_name)
    print(bobbi_list)
        
    harsha_list = []
    for bobbi in harsha_folder:
        bobbi_elec = os.path.basename(bobbi)
        bobbi_elec_name = bobbi_elec.split('_comp')[0]
        harsha_list.append(bobbi_elec_name)
    print(harsha_list)
```

### Generating p-values for Permutations

```python
shuffle_folder = '20210401-phase-shuffle'
no_shuffle_folder = '20210401-no-shuffle'
person_flag = 'harsha'

# shuffle_folder = 'bobbi-matlab-perm'
# no_shuffle_folder = 'bobbi-matlab-no-perm'
# person_flag = 'bobbi'

harsha_elec_sort_key = lambda x: os.path.splitext(os.path.basename(x))[0].split('_comp')[0]
bobbi_noshuffle_elec_sort_key = lambda x: os.path.splitext(os.path.basename(x))[0].split('conversation1')[-1].split('.')[0]
bobbi_shuffle_elec_sort_key = lambda x: os.path.splitext(os.path.basename(x))[0].split('_perm')[0]

shuffle_sort_key = {'harsha': harsha_elec_sort_key, 'bobbi': bobbi_shuffle_elec_sort_key}
main_sort_key = {'harsha': harsha_elec_sort_key, 'bobbi': bobbi_noshuffle_elec_sort_key}

output_file_name = 'jupyter_' + person_flag + '.csv'

shuffle_subjects = sorted(glob.glob(os.path.join(parent_dir, shuffle_folder, '*')))
noshuffle_subjects = sorted(glob.glob(os.path.join(parent_dir, no_shuffle_folder, '*')))

hemisphere_indicator = load_pickle('/scratch/gpfs/hgazula/podcast_hemisphere_indicator.pkl')

lags = np.arange(-2000, 2001, 25)

pval_dict = dict()
some_list = []
for shuffle_subject, noshuffle_subject in zip(shuffle_subjects, noshuffle_subjects):
    if person_flag == 'harsha':
        subject_key = os.path.basename(shuffle_subject)
    elif person_flag == 'bobbi':
        match = re.search('NY(\d+)', os.path.basename(shuffle_subject))
        subject_key = match.group(1)
    else:
        print('invalid person flag')
    
    shuffle_elec_file_list = glob.glob(os.path.join(shuffle_subject, '*.csv'))
    
    if not shuffle_elec_file_list:
        shuffle_elec_file_list = glob.glob(os.path.join(shuffle_subject, '*.mat'))
        
    shuffle_elec_file_list = sorted(shuffle_elec_file_list, key=shuffle_sort_key[person_flag])
    
    main_elec_file_list = glob.glob(os.path.join(noshuffle_subject, '*.csv'))
    
    if not main_elec_file_list:
        main_elec_file_list = glob.glob(os.path.join(noshuffle_subject, '*.mat'))
        
    main_elec_file_list = sorted(main_elec_file_list, key=main_sort_key[person_flag])
    
    a = [os.path.basename(item) for item in shuffle_elec_file_list]
    b = [os.path.basename(item) for item in main_elec_file_list]
    
    assert len(a) == len(b), "Mismatch: Electrode Set"

    for elec_file1, elec_file2 in zip(shuffle_elec_file_list,
                                      main_elec_file_list):

        if person_flag == 'harsha':
            elecname1 = harsha_elec_sort_key(elec_file1)
            elecname2 = harsha_elec_sort_key(elec_file2)
        elif person_flag == 'bobbi':
            elecname1 = bobbi_shuffle_elec_sort_key(elec_file1)
            elecname2 = bobbi_noshuffle_elec_sort_key(elec_file2)
            
        assert elecname1 == elecname2, 'Mismatch: Electrode Name'

        if elecname1.startswith(('SG', 'ECGEKG', 'EEGSG')):
            continue

        if os.path.splitext(elec_file1)[-1] == '.csv':
            perm_result = pd.read_csv(elec_file1, header=None).values
        elif os.path.splitext(elec_file1)[-1] == '.mat':
            perm_result = loadmat(elec_file1)['rc'][0]
        else:
            print('invalid file format')

        if os.path.splitext(elec_file2)[-1] == '.csv':
            rc_result = pd.read_csv(elec_file2, header=None).values
        elif os.path.splitext(elec_file2)[-1] == '.mat':
            rc_result = loadmat(elec_file2)['rc'][0]
        else:
            print('invalid file format')
        
        perm_result = perm_result.reshape(-1, len(lags))
        rc_result = rc_result.reshape(1, -1)
        
        if rc_result.shape[1] == 801:
            rc_result = rc_result[:, 320:-320]
            
        perm_result = perm_result[:1000, :]
        
        assert perm_result.shape[1] == rc_result.shape[
            1], "Mismatch: Number of Lags"

        if perm_result.shape[1] != len(lags):
            print('perm is wrong length')
        else:
            omaxs = np.max(perm_result, axis=1)

        s = 1 - (sum(np.max(rc_result) > omaxs) / perm_result.shape[0])
        some_list.append((subject_key, elecname1, s))

df = pd.DataFrame(some_list, columns=['subject', 'electrode', 'score'])
thresh = 0.01

df1 = df.copy(deep=True)
pre_flag = np.logical_or(np.isclose(df1.score.values, thresh, atol=1e-6), df1.score.values > thresh)

_, pcor, _, _ = multitest.multipletests(df.score.values,
                                        method='fdr_bh',
                                        is_sorted=False)

post_flag = np.logical_or(np.isclose(pcor, thresh), pcor < thresh)

df['pre_flag'] = pre_flag
df['post_flag'] = post_flag
df['pcor'] = pcor

df.to_csv(output_file_name, index=False)

df = df[post_flag]
df.to_csv('selected_' + output_file_name, index=False)

filter_hemisphere = []
for row in df.itertuples(index=False):
    subject = row.subject
    electrode = row.electrode

    curr_key = hemisphere_indicator.get(int(subject), None)

    if not curr_key:
        if int(subject) == 798:
            filter_hemisphere.append((subject, electrode))    
    elif len(curr_key) == 2:
        if electrode.startswith(('L', 'DL')):
            filter_hemisphere.append((subject, electrode))
    elif len(curr_key) == 1 and 'RH' in curr_key:
        continue
    else:
        filter_hemisphere.append((subject, electrode))
        
df2 = pd.DataFrame(filter_hemisphere, columns=['subject', 'electrode'])
df2.to_csv('selected_lh_' + output_file_name, index=False, columns=['subject', 'electrode'])
```

### Stability of Bobbi's Permutations

```python
b1 = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_bobbi_1000.csv'
b2 = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_bobbi_2000.csv'
b3 = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_bobbi_3000.csv'
b4 = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_bobbi_4000.csv'
b5 = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_bobbi_5000.csv'
h = '/scratch/gpfs/hgazula/podcast-encoding/results/selected_lh_jupyter_harsha.csv'

b1 = pd.read_csv(b1)
b2 = pd.read_csv(b2)
b3 = pd.read_csv(b3)
b4 = pd.read_csv(b4)
b5 = pd.read_csv(b5)
h = pd.read_csv(h)

b1 = b1[['subject', 'electrode']].values.tolist()
b2 = b2[['subject', 'electrode']].values.tolist()
b3 = b3[['subject', 'electrode']].values.tolist()
b4 = b4[['subject', 'electrode']].values.tolist()
b5 = b5[['subject', 'electrode']].values.tolist()
h = h[['subject', 'electrode']].values.tolist()

print(f'b1: {len(b1)}')
print(f'b2: {len(b2)}')
print(f'b3: {len(b3)}')
print(f'b4: {len(b4)}')
print(f'b5: {len(b5)}')
print(f'h: {len(h)}')
print()

b1 = ['_'.join([str(elem) for elem in item]) for item in b1]
b2 = ['_'.join([str(elem) for elem in item]) for item in b2]
b3 = ['_'.join([str(elem) for elem in item]) for item in b3]
b4 = ['_'.join([str(elem) for elem in item]) for item in b4]
b5 = ['_'.join([str(elem) for elem in item]) for item in b5]
h = ['_'.join([str(elem) for elem in item]) for item in h]

# Not in harsha
nih1 = sorted(set(b1) - set(h))
nih2 = sorted(set(b2) - set(h))
nih3 = sorted(set(b3) - set(h))
nih4 = sorted(set(b4) - set(h))
nih5 = sorted(set(b5) - set(h))

print(f'not in harsha1: {nih1}')
print(f'not in harsha2: {nih2}')
print(f'not in harsha3: {nih3}')
print(f'not in harsha4: {nih4}')
print(f'not in harsha5: {nih5}')
print()

l1 = [(item.split('_')[0], item.split('_')[1]) for item in nih1]
l2 = [(item.split('_')[0], item.split('_')[1]) for item in nih2]
l3 = [(item.split('_')[0], item.split('_')[1]) for item in nih3]
l4 = [(item.split('_')[0], item.split('_')[1]) for item in nih4]
l5 = [(item.split('_')[0], item.split('_')[1]) for item in nih5]

pd.DataFrame(l1, columns=['subject', 'electrode']).to_csv('notinharsha1.csv', index=False)
pd.DataFrame(l2, columns=['subject', 'electrode']).to_csv('notinharsha2.csv', index=False)
pd.DataFrame(l3, columns=['subject', 'electrode']).to_csv('notinharsha3.csv', index=False)
pd.DataFrame(l4, columns=['subject', 'electrode']).to_csv('notinharsha4.csv', index=False)
pd.DataFrame(l5, columns=['subject', 'electrode']).to_csv('notinharsha5.csv', index=False)

# Not in bobbi
nib1 = sorted(set(h) - set(b1))
nib2 = sorted(set(h) - set(b2))
nib3 = sorted(set(h) - set(b3))
nib4 = sorted(set(h) - set(b4))
nib5 = sorted(set(h) - set(b5))

print(f'not in bobbi1: {nib1}')
print(f'not in bobbi2: {nib2}')
print(f'not in bobbi3: {nib3}')
print(f'not in bobbi4: {nib4}')
print(f'not in bobbi5: {nib5}')

l1 = [(item.split('_')[0], item.split('_')[1]) for item in nib1]
l2 = [(item.split('_')[0], item.split('_')[1]) for item in nib2]
l3 = [(item.split('_')[0], item.split('_')[1]) for item in nib3]
l4 = [(item.split('_')[0], item.split('_')[1]) for item in nib4]
l5 = [(item.split('_')[0], item.split('_')[1]) for item in nib5]

pd.DataFrame(l1, columns=['subject', 'electrode']).to_csv('notinbobbi1.csv', index=False)
pd.DataFrame(l2, columns=['subject', 'electrode']).to_csv('notinbobbi2.csv', index=False)
pd.DataFrame(l3, columns=['subject', 'electrode']).to_csv('notinbobbi3.csv', index=False)
pd.DataFrame(l4, columns=['subject', 'electrode']).to_csv('notinbobbi4.csv', index=False)
pd.DataFrame(l5, columns=['subject', 'electrode']).to_csv('notinbobbi5.csv', index=False)
```
